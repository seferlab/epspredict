```python
def adjust_to_quarter(date):
    year = date.year
    if date.month in [1, 2]:
        return pd.Timestamp(year=year-1, month=12, day=31)
    elif date.month in [4, 5]:
        return pd.Timestamp(year=year, month=3, day=31)
    elif date.month in [7, 8]:
        return pd.Timestamp(year=year, month=6, day=30)
    elif date.month in [10, 11]:
        return pd.Timestamp(year=year, month=9, day=30)
    else:
        return date
```

```python
def merge_tables(table1, table2):
    table1["Adjusted Date"] = pd.to_datetime(table1['Report Date']).apply(adjust_to_quarter)
    
    table1['Year'] = pd.to_datetime(table1['Adjusted Date']).dt.year
    table1['Month'] = pd.to_datetime(table1['Adjusted Date']).dt.month

    table2['Year'] = pd.to_datetime(table2['PENDS']).dt.year
    table2['Month'] = pd.to_datetime(table2['PENDS']).dt.month
    
    merged_table = pd.merge(table1, table2, on=['OFTIC', 'Year', 'Month'], how='inner')
    
    return merged_table
```

```python
merged_data = merge_tables(sec_data, data)
```
